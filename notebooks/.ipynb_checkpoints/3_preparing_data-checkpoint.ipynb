```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_duration = pd.read_csv('../data/processed/df_duration.csv',index_col=0)
```

```python
df_duration.drop('Unnamed: 0.1',axis = 1, inplace = True)
```

```python
df_duration
```

```python
df_duration['publish_time'] = pd.to_datetime(df_duration.publish_time)
```

```python
df_duration['publish_time'] = df_duration.publish_time.dt.date
```

```python
df_duration
```

```python
df_duration['trending_date'] = pd.to_datetime(df_duration.trending_date,format = "%y.%d.%m")
```

```python
df_duration
```

```python
df_duration.sort_values(by = 'publish_time',axis = 0,ignore_index=True,inplace = True)
```

```python
df_duration.head()
```

```python
df_duration.info()
```

```python
df_duration['publish_time'] = df_duration.publish_time.astype('datetime64[ns]')
```

```python
df_duration.info()
```

```python
from datetime import date
import calendar
# curr_date = date.today()
df_duration['day_of_week'] = df_duration['publish_time'].apply(lambda x : calendar.day_name[x.weekday()])
```

```python
df_duration['day_of_week']
```

```python
df_duration['published_in_weekend'] = [True if day in ['Saturday','Sunday'] else False for day in df_duration['day_of_week']]
```

```python
df_duration[df_duration['published_in_weekend'] == True]
```

```python
df_duration.rename(columns = {'day_of_week':'puplished_day'},inplace = True)
```

```python
df_duration.info()
```

```python
df_duration = df_duration[df_duration.duration_in_minutes > 0]
```

```python
df_duration
```

```python
df_duration.to_csv('../data/processed/df_duration_prepared.csv')
```
