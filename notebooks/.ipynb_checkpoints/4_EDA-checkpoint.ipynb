```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
```

```python
df = pd.read_csv('../data/processed/df_duration_prepared.csv',index_col=0)
```

```python
days = df.puplished_day.value_counts().sort_values()
puplished_day_views_norm = df.groupby('puplished_day')['views'].sum()[days.index] / days
```

```python
px.bar(df,y = days.index,x = days,color_continuous_scale='orrd',
       orientation='h',color = puplished_day_views_norm)
```
