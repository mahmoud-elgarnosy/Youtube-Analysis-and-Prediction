```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('../data/external/final_data/final_data_without_duplicates_10_elbagory_scrapping.csv',index_col=0)
```

```python
df1 = df1[~df1.duration.isnull()]
```

```python
df1
```

```python
#there are some errors in reading duration 
erorrs = df1[df1.duration.apply(lambda x : x[-2:].count('0') == 2 and x.count(':') == 2)]
erorrs
```

```python
errd = erorrs.duration.apply(lambda x : x[:-3])
errd
```

```python
erorrs.drop('duration',axis = 1,inplace =True)
```

```python
erorrs['duration'] =  errd
```

```python
erorrs
```

```python
df1 = df1[~df1['video_id'].isin(list(erorrs.video_id))]
```

```python
df1 = df1.append(erorrs,ignore_index = True)
```

```python
df1
```

```python
df2 = pd.read_csv('../data/external/final_data/final_data_without_duplicates_10_mahmoud_scrapping.csv')
df2 = df2[~df2.duration.isnull()]
df2
```

```python
df_duration = df1.append(df2,ignore_index=True)
```

```python
df_duration
```

```python
df_duration.drop_duplicates(subset='video_id',inplace = True, ignore_index=True)
```

```python
df_duration.info()
```

```python
import time
def convert(time_string):
    if time_string.count(':') == 2:
        date_var = time.strptime(time_string, '%H:%M:%S')

    else:    
        date_var = time.strptime(time_string, '%M:%S')

    minutes = date_var.tm_hour * 60 + date_var.tm_min
    
    if date_var.tm_sec > 30:
        minutes+=1
    return minutes
```

```python
duration_in_minutes = df_duration.duration.apply(lambda x : convert(x))
duration_in_minutes
```

```python
df_duration['duration_in_minutes'] = duration_in_minutes
```

```python
df_duration
```

```python
df_duration.drop('duration',inplace=True,axis = 1)
```
